```python
# 1) MonthlyIncome Review;

df[df['MonthlyIncome'].isnull()]
```

```python
# MonthlyIncome is most highly related with three columns: JobLevel, TotalWorkingYears, and Age.

df[["MonthlyIncome", "JobLevel", "TotalWorkingYears", "Age"]].corr()
```

```python
#Calculate the average monthly income

df.MonthlyIncome.mean()
```

```python
# Calculate the Average of MonthlyIncome column by 'JobLevel','TotalWorkingYears'

df.groupby(['JobLevel','TotalWorkingYears'])['MonthlyIncome'].transform('mean')
```

```python
# Filling the missing values of MonthlyInCome column by the Calculated Average of MonthlyIncome

df['MonthlyIncome'].fillna(df.groupby(['JobLevel','TotalWorkingYears'])['MonthlyIncome'].transform('mean'),inplace=True)
```

```python
df[['MonthlyIncome','JobLevel','TotalWorkingYears']].head()   # checking the Null values on MonthlyIncome
```

```python
# 2) Department Review;

df[df['Department'].isnull()]
```

```python
# Department column is most highly related with the column 'JobRole'.

df.groupby('Department')['JobRole'].value_counts()
```

```python
# Fill the missing values of the Department column by the mode() of 'JobRole' column

df['Department'].fillna(df.groupby(["JobRole"])["Department"].transform(lambda x: x.mode()[0]),inplace=True)
```

```python
df['Department'].isnull().sum() # Checking the Department columns's Null values
```

```python
df.iloc[[7]]
```

```python
# 3) EducationField Review;

df[df['EducationField'].isnull()]
```

```python
# EducationField column is most highly related with the columns; Department, TotalWorkingYears or JobRole.

df.groupby(["Department","MonthlyIncome"])["EducationField"].value_counts()
```

```python
# Fill the missing values of EducationField column by the mode of "Department" and "TotalWorkingYears" column

df.EducationField.fillna(df.groupby(["Department","TotalWorkingYears"])["EducationField"].transform(lambda x: x.mode()[0]),inplace=True)
```

```python
# 4) EmployeeNumber Review;

df[df['EmployeeNumber'].isnull()]
```

```python
df['EmployeeNumber'].nunique()
```

```python
df['EmployeeNumber'].agg([max , min])
```

```python
df['EmployeeNumber'].agg([max , min])
```

```python
# Fill the missing values of EmployeeNumber column

df['EmployeeNumber'].fillna(df['EmployeeNumber'].max()+1,inplace=True)
```

```python
# 5) Gender Review;

df[df['Gender'].isnull()]
```

```python
df.groupby(["Department","Gender","MaritalStatus"])["MonthlyIncome"].mean()
```

```python
# Fill the missing values of Gender column

df['Gender'].fillna(df.groupby(["Department","MaritalStatus"])["Gender"].transform(lambda x: x.mode()[0]),inplace=True)
```

```python
# 6) MaritalStatus Review;

df[df['MaritalStatus'].isnull()]
```

```python
# Fill the missing values of MaritalStatus column

df['MaritalStatus'].fillna(df.groupby(["Department","Gender"])["MaritalStatus"].transform(lambda x: x.mode()[0]),inplace=True)
```

```python
# 7) JobRole Review;

df[df['JobRole'].isnull()]
```

```python
df.groupby(["Department","JobRole","EducationField","JobLevel"])["Gender"].value_counts(normalize=True)*100
```

```python
# Fill the missing values of MaritalStatus column
df['JobRole'].fillna(df.groupby(["Department","EducationField","Gender","JobLevel"])["JobRole"].transform(lambda x: x.mode()[0]),inplace=True)
```

```python
# 8) MonthlyRate Review

df[df['MonthlyRate'].isnull()]
```

```python
# Fill the missing values of MonthlyRate column

df.groupby(["Department", "JobRole", "EducationField",
            "Gender"])["MonthlyRate"].transform(
                lambda x: x.mode().iloc[0] if not x.mode().empty else x.mean())
```

```python
# 9) PercentSalaryHike Review;

df[df['PercentSalaryHike'].isnull()]
```

```python
# Fill the missing values of PercentSalaryHike column

df.PercentSalaryHike.fillna(df.groupby(["Department","JobRole","Age"])["PercentSalaryHike"].transform(lambda x: x.mode()[0]),inplace=True)
```

## <font color='dodgerblue'> <b> PANDAS WITH FUNCTIONS </b><font color='black'>

```python
# Let's read a CSV file using Pandas as follows
hr_df = pd.read_csv('Human_Resources.csv')
hr_df
```

- **Let's assume the daily rate has increased by 10%**
- **Define a function that increases all clients networth by a fixed value of 10% (for simplicity sake)**
- **assume that net worth increased by 10%**
- **You can apply a function to the DataFrame**

```python
#Let's assume the daily rate has increased by 10%
#Define a function that increases all clients networth by a fixed value of 10% (for simplicity sake)

def increase_by_10_percent(value):
    return value * 1.10
```

```python
#assume that net worth increased by 10%
hr_df['UpdatedDailyRate'] = hr_df['DailyRate'].apply(increase_by_10_percent)
```

```python
#You can apply a function to the DataFrame
hr_df.head()
```

📍**PRACTICE OPPORTUNITY #7:**
- **Define a function that doubles the DailyRate and adds $100**
- **Apply the function to the DataFrame**
- **Calculate the updated total Daily Rate of all employees combined**

```python
# Define a function that doubles the DailyRate and adds $100 
def double_and_add(daily_rate):
    return daily_rate * 2 + 100
```

```python
#Apply the function to the DataFrame
hr_df['UpdatedDailyRate2'] = hr_df['DailyRate'].apply(double_and_add)
```

```python
#Calculate the updated total Daily Rate of all employees combined
total_updated_daily_rate = hr_df['UpdatedDailyRate2'].sum()
```

```python
print("Updated Daily Rate2:", total_updated_daily_rate)
```
