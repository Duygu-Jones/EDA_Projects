<h3 align="left"><font color=dodgerblue>  Getting to Know About Cancer</font></h3>  
    
  

*Cancer is a disease characterized by the uncontrolled growth and spread of abnormal cells in the body. These abnormal cells can invade and destroy normal body tissue and can spread to other parts of the body (metastasize). Cancer can affect almost any part of the body and has the potential to be life-threatening if not diagnosed and treated promptly. The causes of cancer are varied and can include genetic factors, lifestyle choices, environmental exposures, and infections. Treatment options for cancer typically include surgery, chemotherapy, radiation therapy, immunotherapy, and targeted therapy, depending on the type and stage of the cancer.*

##  <font color=#2865c8> <b> DATA VISUALIZATION </b><font color='black'> 


<h3 align="left"><font color=dodgerblue> Plot & ScatterPlot & CountPlot using Seaborn </font></h3> 

**About Seaborn and Matplotlib Libraries**

- **Seaborn is a visualization library that sits on top of matplotlib**
- **Seaborn offers enhanced features compared to matplotlib**
- **https://seaborn.pydata.org/examples/index.html*/*


<h3 align="left"><font color=dodgerblue> Tasks-2 </font></h3>
   
- **Plot scatter plot between mean area and mean smoothness**
- **Let's print out countplot to know how many samples belong to class #0 and #1**

```python
#Plot scatter plot between mean area and mean smoothness

from matplotlib.patches import Patch

sns.scatterplot(data=df, x='mean area', y='mean smoothness', hue='target',palette={1: 'blue', 0: 'red'})
plt.title('Mean Area vs Mean Smoothness')

legend_labels = ['1: Benign', '0: Malignant']
legend_colors = ['blue', 'red']
legend_handles = [Patch(facecolor=color, label=label) for color, label in zip(legend_colors, legend_labels)]
plt.legend(handles=legend_handles, title='Diagnosis')
plt.show()
```

This scaltter plot..........

```python
# Let's print out countplot to know how many samples belong to class #0 and #1

ax = sns.countplot(data=df, x='target',palette={1: 'blue', 0: 'red'})
plt.title('Count of Samples for Each Class')
ax.bar_label(ax.containers[0])
plt.show()
```

### <font color='dodgerblue'> <b> PLOT PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS USING SEABORN</b><font color='orange'>

<h3 align="left"><font color=dodgerblue> Tasks-3 </font></h3>
   
- **Plot the pairplot**
- **Strong correlation between the mean radius and mean perimeter, mean area and mean primeter**
- **plot the distplot** 
- **Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)**
- **KDE is used to plot the Probability Density of a continuous variable.** 

```python
# Plot the pairplot Strong correlation between the mean radius and mean perimeter, mean area and mean primeter

sns.pairplot(
    df,
    vars=['mean radius', 'mean perimeter', 'mean area', 'mean smoothness'],
    hue='target',
    palette={
        1: 'blue',
        0: 'red'
    })

plt.show()

# Yorum: Bu plot mean_radius, mean_perimeter ve mean_area arasnda gl korelasyonlar olarak gzlemlenebilir.
```

<div style="border-radius:10px;border:#254E58 solid;padding: 15px;background-color:white;font-size:110%;text-align:left">

<h4 align="left"><font color=dodgerblue>  Output: Scatterplot for Mean Area vs Mean Smoothness:</font></h4>   
 
This scaltter plot..........

```python
# plot the distplot Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)

sns.distplot(df['mean radius'], kde=True)
plt.title('Distribution of Mean Radius')
plt.show()
```
