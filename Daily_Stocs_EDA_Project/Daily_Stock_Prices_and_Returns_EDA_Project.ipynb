```python
# 5. Using Matplotlib, plot the 4 plots on subplots next to each other (all figures in one row).

# Create a figure and a set of subplots
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(20, 5), sharex=True)

# Plot each stock's daily price on separate subplots

# AAPL
ax[0].plot(df_prices.index, df_prices['AAPL'], label='AAPL')
ax[0].set_title('Daily Prices of AAPL')
ax[0].set_ylabel('Price')
ax[0].legend()
ax[0].grid(True)

# JPM
ax[1].plot(df_prices.index, df_prices['JPM'], label='JPM')
ax[1].set_title('Daily Prices of JPM')
ax[1].set_ylabel('Price')
ax[1].legend()
ax[1].grid(True)

# PG
ax[2].plot(df_prices.index, df_prices['PG'], label='PG')
ax[2].set_title('Daily Prices of PG')
ax[2].set_ylabel('Price')
ax[2].legend()
ax[2].grid(True)

# UAL
ax[3].plot(df_prices.index, df_prices['UAL'], label='UAL')
ax[3].set_title('Daily Prices of UAL')
ax[3].set_ylabel('Price')
ax[3].legend()
ax[3].grid(True)

# Add a common x-axis label
fig.text(0.5, 0.04, 'Date', ha='center')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

<font color='dodgerblue'> <b> Bonus: Scatter Plot</b><font color='orange'>

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px


# Veri setini yeniden ykleyelim
returns_df = pd.read_csv('stocks_daily_returns.csv')
returns_df['Date'] = pd.to_datetime(returns_df['Date'])
returns_df.set_index('Date', inplace=True)

# Create a DataFrame for plotting with plotly
plot_df = returns_df.reset_index()

# Add formatted date column for hover data
plot_df['Formatted Date'] = plot_df['Date'].dt.strftime('%Y-%m-%d')

# Create the figure
fig = go.Figure()

# Add Apple scatter plot
fig.add_trace(go.Scatter(
    x=plot_df['AAPL'],
    y=plot_df['JPM'],
    mode='markers',
    marker=dict(color='blue', size=5),
    text=plot_df['Formatted Date'],
    name='Apple',
    hovertemplate='Date: %{text}<br>Apple Returns: %{x}<br>'
))

# Add JP Morgan scatter plot
fig.add_trace(go.Scatter(
    x=plot_df['JPM'],
    y=plot_df['AAPL'],
    mode='markers',
    marker=dict(color='saddlebrown', size=5),
    text=plot_df['Formatted Date'],
    name='JP Morgan',
    hovertemplate='Date: %{text}<br>JP Morgan Returns: %{x}<br>'
))

# Set the titles and labels
fig.update_layout(
    title='Scatterplot of Daily Returns: Apple vs. JP Morgan',
    xaxis_title='Apple Daily Returns',
    yaxis_title='JP Morgan Daily Returns',
    hovermode='closest'
)

# Show the plot
fig.show()
```

```python
#7) Using Seaborn, plot a similar scatterplot between Apple and JP Morgan daily returns.

# Create a scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_returns['AAPL'], y=df_returns['JPM'], alpha=0.5)
plt.title('Scatter Plot between Apple and JP Morgan Daily Returns')
plt.xlabel('Apple Daily Returns')
plt.ylabel('JP Morgan Daily Returns')
plt.grid(True)

# Save the plot as an image file
plt.savefig('seaborn_scatter_plot.png', dpi=300)

# Show plot
plt.show()
```

### <font color='dodgerblue'> <b>Pie Chart</b><font color='orange'>

```python
# 8) Using Matplotlib, plot a pie chart that shows asset allocations, using the 'explode' attribute to 
#increase the separation between AAPL and the rest of the portfolio.

# Define the asset allocations
allocations = [40, 25, 20, 15]  # Example percentages for AAPL, JPM, PG, UAL
labels = ['AAPL', 'JPM', 'PG', 'UAL']

# Define the explode parameter to separate AAPL from the rest
explode = [0.1, 0, 0, 0]

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(allocations, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Asset Allocations')
plt.show()
```

<font color=dodgerblue>  Output: Pie Chart</font>   
 
This pie chart shows the asset allocation in the portfolio. AAPL holds the largest share at 40%, followed by JPM at 25%, PG at 20%, and UAL at 15%. AAPL is separated from the rest using the 'explode' feature, indicating the investor's high confidence and high return expectation in Apple. The chart visually presents the diversification and the risk-return balance in the portfolio.

```python
# 9. Assume that you decided to become bullish on AAPL and you allocated 70% of your assets in it. 
# You also decided to equally divide the rest of your assets in other stocks (JPM, PG, and UAL). 
# Using Matplotlib, plot a pie chart that shows these allocations. 
# Use 'explode attribute to increase the separation between AAPL and the rest of the portfolio.

# Define the new asset allocations
allocations = [70, 10, 10, 10]  # New percentages for AAPL, JPM, PG, UAL
labels = ['AAPL', 'JPM', 'PG', 'UAL']

# Define the explode parameter to separate AAPL from the rest
explode = [0.1, 0, 0, 0]

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(allocations, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140,colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('New Asset Allocations')
plt.show()
```

### <font color='dodgerblue'> <b> Histogram Graph</b><font color='orange'>

```python
# 10. Using Matplotlib, plot the histogram for United Airlines and P&G returns using 40 bins with red color. 
# Display the mean and standard deviation for both stocks on top of the figure.

# Calculate the mean and standard deviation for UAL and PG returns
ual_mean = df_returns['UAL'].mean()
ual_std = df_returns['UAL'].std()
pg_mean = df_returns['PG'].mean()
pg_std = df_returns['PG'].std()

# Plot the histograms for UAL and PG returns
plt.figure(figsize=(10, 5))

# Histogram for UAL
plt.subplot(1, 2, 1)
plt.hist(df_returns['UAL'], bins=40, color='red', alpha=0.7)
plt.title(f'UAL Returns\nMean: {ual_mean:.2f}, Std: {ual_std:.2f}')
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.grid(True)

# Histogram for PG
plt.subplot(1, 2, 2)
plt.hist(df_returns['PG'], bins=40, color='red', alpha=0.7)
plt.title(f'PG Returns\nMean: {pg_mean:.2f}, Std: {pg_std:.2f}')
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

11. Using Seaborn, plot a heatmap that shows the correlations between stocks daily returns. Comment on the correlation between UAL and P&G.
    

```python
# Calculate the correlation matrix
correlation_matrix = df_returns.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Stocks Daily Returns')
plt.show()

# Comment on the correlation between UAL and PG
ual_pg_correlation = correlation_matrix.loc['UAL', 'PG']
print(f'The correlation between UAL and PG is: {ual_pg_correlation:.4f}')
```

### <font color='dodgerblue'> <b> 3D Plot</b><font color='orange'>

<font color=dodgerblue> Tasks-4 </font>   

12. Plot a 3D plot showing all daily returns from JPM, AAPL, and UAL.
    


```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the datasets
df_returns = pd.read_csv('stocks_daily_returns.csv')

# Convert 'Date' column to datetime format
df_returns['Date'] = pd.to_datetime(df_returns['Date'])

# Set 'Date' column as the index (optional)
df_returns.set_index('Date', inplace=True)

# Extracting the relevant data
jpm_returns = df_returns['JPM']
aapl_returns = df_returns['AAPL']
ual_returns = df_returns['UAL']

# Enable interactive plots
%matplotlib notebook

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.scatter(jpm_returns, aapl_returns, ual_returns, c='r', marker='o', alpha=0.6)

# Add labels
ax.set_xlabel('JP Morgan Daily Returns')
ax.set_ylabel('Apple Daily Returns')
ax.set_zlabel('United Airlines Daily Returns')
ax.set_title('3D Plot of Daily Returns: JPM, AAPL, and UAL')

# Show the plot
plt.show()
```
