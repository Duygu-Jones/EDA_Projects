## <font color=#2865c8> <b> Let's fill in the missing data according to our correlation analysis </b><font color='orange'>

```python
# MonthlyIncome Review;

df[df['MonthlyIncome'].isnull()]
```

```python
#Calculate the average monthly income
df.MonthlyIncome.mean()
```

```python
# MonthlyIncome is most highly related with three columns: JobLevel, TotalWorkingYears, and Age.

df[["MonthlyIncome", "JobLevel", "TotalWorkingYears", "Age"]].corr()
```

```python
# Filling the missing values of MonthlyInCome column by the Calculated Average of MonthlyIncome

df['MonthlyIncome'].fillna(df.groupby(['JobLevel','TotalWorkingYears'])['MonthlyIncome'].transform('mean'),inplace=True)
```

```python
df[['MonthlyIncome','JobLevel','TotalWorkingYears']].head(8)   # checking the Null values on MonthlyIncome
```

```python
# Department Review;

df[df['Department'].isnull()]
```

```python
# Department column is most highly related with the column 'JobRole'.

df.groupby(['Department','EducationField'])['JobRole'].value_counts()
```

```python
# Fill the missing values of the Department column by the mode() of 'JobRole' column

df['Department'].fillna(df.groupby(["JobRole",'EducationField'])["Department"].transform(lambda x: x.mode()[0]),inplace=True)
```

```python
df['Department'].isnull().sum() # Checking the Department columns's Null values
```

```python
# EducationField Review;

df[df['EducationField'].isnull()]
```

```python
# EducationField column is most highly related with the columns; Department, TotalWorkingYears or JobRole.

df.groupby(["Department","MonthlyIncome"])["EducationField"].value_counts().head(8)
```

```python
# Fill the missing values of EducationField column by the mode of "Department" and "TotalWorkingYears" column

df.EducationField.fillna(df.groupby(["Department","TotalWorkingYears"])["EducationField"].transform(lambda x: x.mode()[0]),inplace=True)
```

```python
# EmployeeNumber Review;

df[df['EmployeeNumber'].isnull()]
```

```python
# Fill the missing values of Gender column

df['Gender'].fillna(df.groupby(["Department","MaritalStatus"])["Gender"].transform(lambda x: x.mode()[0]),inplace=True)
```

# <font color=#2865c8> <b> DEPARTMENT AND JOB ROLE ANALYSIS </b><font color='black'>

## <font color=#2865c8> <b> Which department has a higher attrition rate? </b><font color='black'>

```python
pyo.init_notebook_mode(connected=True)

# Prepare Department and Attrition data
department_attrition = pd.crosstab(df['Department'], df['Attrition'], normalize='index').reset_index()

# Rename the columns
department_attrition.columns = ['Department', 'No', 'Yes']

# Yüzde formatını hesaplayalım
department_attrition['No'] = department_attrition['No'] * 100
department_attrition['Yes'] = department_attrition['Yes'] * 100

# Create an interactive bar chart using Plotly
fig = px.bar(department_attrition, 
             x='Department', 
             y=['No', 'Yes'], 
             barmode='stack', 
             labels={'value': 'Proportion (%)', 'Department': 'Department'},
             title='Department and Attrition')

# Set the content of the tooltip
fig.update_traces(hovertemplate='%{x}<br>Proportion: %{y:.2f}%<extra></extra>', texttemplate='%{y:.2f}%', textposition='inside')

# Customize the chart layout
fig.update_layout(xaxis_title='Department', yaxis_title='Proportion (%)', title='Department and Attrition')
fig.show()

```

<div style="border-radius:10px;border:#254E58 solid;padding: 15px;background-color:white;font-size:110%;text-align:left">

<font color=dodgerblue> 💡 Output: Graph Analysis:</font>

## <font color=#2865c8> <b>Which job role has a higher attrition rate?</b><font color='black'>

```python
# Prepare Job Role and Attrition data
job_role_attrition = pd.crosstab(df['JobRole'], df['Attrition'], normalize='index').reset_index()
job_role_attrition.columns = ['JobRole', 'No', 'Yes']
job_role_attrition['No'] *= 100
job_role_attrition['Yes'] *= 100

# Add department information for each job role
job_role_departments = {
    'Sales Executive': 'Sales',
    'Research Scientist': 'Research & Development',
    'Laboratory Technician': 'Research & Development',
    'Manufacturing Director': 'Research & Development',
    'Healthcare Representative': 'Sales',
    'Manager': 'Research & Development',
    'Sales Representative': 'Sales',
    'Research Director': 'Research & Development',
    'Human Resources': 'Human Resources'
}
job_role_attrition['Department'] = job_role_attrition['JobRole'].map(job_role_departments)

# Create an interactive bar chart using Plotly
fig = px.bar(job_role_attrition, 
             x='JobRole', 
             y=['No', 'Yes'], 
             barmode='stack', 
             labels={'value': 'Proportion (%)', 'JobRole': 'Job Role'},
             title='Job Role and Attrition',
             custom_data=['Department'])

# Set the content of the tooltip
fig.update_traces(hovertemplate='Department: %{customdata[0]}<br>Job Role: %{x}<br>Proportion: %{y:.2f}%<extra></extra>')

# Customize the chart layout
fig.update_layout(xaxis_title='Job Role', yaxis_title='Proportion (%)')
fig.show()
```

## <font color=#2865c8> <b> What are the attrition rates by age groups?</b><font color='black'>

```python
# Define age groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[18, 30, 40, 50, 60], labels=['18-30', '30-40', '40-50', '50+'], right=False)

# Calculate attrition rates
attrition_summary = df[df['Attrition'] == 'Yes']['AgeGroup'].value_counts() / df['AgeGroup'].value_counts() * 100

# Create bar chart
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(attrition_summary.index, attrition_summary.values)

# Add percentages on top of bars
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}%',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

plt.title('Attrition Rates by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)
plt.show()
```

Scaling, normalization, and standardization are common preprocessing steps to prepare data for machine learning models. 

- Standardization transforms data to have **mean**:0 and a **standard deviation: 1**.
- Normalization typically scales the values in a column to the range between 0 and 1.


**Use Cases for Standardization**:

- Linear Regression, Logistic Regression, and Neural Networks: These algorithms require features to have the same scale and mean.
- Principal Component Analysis (PCA): Requires features to have the same scale and mean.

**Use Cases for Normalization**:

- Min-Max Scaling in Deep Learning Models: Especially in neural networks, having inputs within a certain range can improve model performance.
- K-Means Clustering: Used to eliminate scale differences among features.

Both methods are important for making data more comparable and improving model performance. However, which method to use depends on the requirements of a specific machine learning algorithm and the characteristics of the dataset.
