<h3 align="left"><font color=#2067d4>Data Overview</font></h3>     
<h4 align="left"><font color=#2067d4> About the Kyphosis Dataset</font></h4>  
    
- Dataset contains 81 rows and 4 columns representing data on children who have had corrective spinal surgery.
<br>    
- INPUTS: 
    1. **Age**: The age of the patient in months., 
    2. **Number**: The number of vertebrae involved in the surgery, 
    3. **Start**: The number of the first (top-most) vertebra operated on.
    4. **Kyphosis**: Indicates whether kyphosis was present (yes/no) after the operation. 

    <br>      
- OUTPUTS: Represents a factor with levels "absent" or "present," indicating if **kyphosis** (a type of deformation) was present after the operation.

**Labels of the Vertebrea Groups**

- (4) Sacrum (S1-S5)

## <font color=#2067d4> <b> Feature Engineering </b><font color='orange'>

### Classify the 'Start' column into Anatomic Vertebrea Groups

```python
# Classify vertebraes into Anatomic groups

def classify_vertebra(start):
    if 1 <= start <= 7:
        return 'Cervical(Neck)'
    elif 8 <= start <= 19:
        return 'Thoracic(Back)'
    elif 20 <= start <= 24:
        return 'Lumbar(Lower Back)'
    elif 25 <= start <= 29:
        return 'Sacral'
    else:
        return 'Unknown'
    
# Apply the function to create a new column
df['Vertebrae Group'] = df['Start'].apply(classify_vertebra)

# Display the unique vertebra groups and their counts
vertebra_groups = df['Vertebrae Group'].value_counts()
print(vertebra_groups)
```

### Label the 'VertebreaGroups' new column's values

```python
# Function to Label the 'VertebreaGroups' values

def classify_vertebra(start):
    if 1 <= start <= 7:
        return 1  # Cervical (Neck)
    elif 8 <= start <= 19:
        return 2  # Thoracic (Back)
    elif 20 <= start <= 24:
        return 3  # Lumbar (Lower Back)
    elif 25 <= start <= 29:
        return 4  # Sacral
    else:
        return 5  # Unknown or other

# Create a new 'VertebreaGroupNumbers' column and apply the classification function
df['Vertebrae Group Number'] = df['Start'].apply(classify_vertebra)

df.head()
```

### Convert the “Age” in months into a new column entitled “Age in Years”

```python
# Function to convert months to years
def convert_months_to_years(months):
    return round(months / 12, 1)

# Apply the function to the 'Age' column
df['Age'] = df['Age'].apply(convert_months_to_years)

df.head()
```

### Rename the Columns

```python
#features in dataset
df.columns
```

```python
# Rename columns
df.rename(columns={
    'Kyphosis': 'Kyphosis Status',
    'Number': 'Number of Vertebrae',
    'Start': 'Start Vertebrae'  
}, inplace=True)
```

```python
df.head()
```

### Convert 'Kyphosis Status' to binary values

```python
# Convert 'Kyphosis Status' to binary values
df['Kyphosis Status'] = df['Kyphosis Status'].map({'present': 1, 'absent': 0})
```

### Calculate the average, minimum, and maximum ages to Analyze the ages of the participants.

```python
# Calculate the average, minimum, and maximum  for age
average_age = df['Age'].mean()
minimum_age = df['Age'].min()
maximum_age = df['Age'].max()

print(f"Average Age: {average_age:.2f} years")
print(f"Minimum Age: {minimum_age:.2f} years")
print(f"Maximum Age: {maximum_age:.2f} years")
```

## <font color=#2067d4> <b> Distributions </b><font color='orange'>

### How is the Distribution of Groups of Vertebrae?

```python
plt.figure(figsize=(6,4))
vertebra_groups.plot(kind='bar', color='skyblue')
plt.title('Distribution of Current Groups of Vertebrae')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Which vertebra group exhibits the highest prevalence of kyphosis and Conversely, which vertebra group shows the lowest prevalence of kyphosis  after the operation? 

```python
# Group and count the kyphosis conditions by vertebraGroup.
kyphosis_by_vertebra = df.groupby('Vertebrae Group')['Kyphosis Status'].value_counts(normalize=True).unstack().fillna(0)

# Visualize Kyphosis Range
plt.figure(figsize=(12, 8))
kyphosis_by_vertebra.plot(kind='bar', stacked=True)
plt.title('Range of Kyphosis by the Vertebrae Groups')
plt.xlabel('Vertebrae Groups')
plt.ylabel('Range')
plt.legend(title='Kyphosis')
plt.show()
```

<font color=dodgerblue> 💡 Output: distribution of kyphosis (presence and absence) across cervical (neck) and thoracic (back) vertebra groups.</font>  

- **Cervical Vertebrae:** Kyphosis is equally distributed, with about 50% presence and 50% absence.
- **Thoracic Vertebrae:** Kyphosis is less common, with approximately 90% absence and only 10% presence.

This suggests that kyphosis is more prevalent in the cervical region compared to the thoracic region.


### How does the presence of kyphosis vary across different vertebra numbers, and which specific vertebra numbers show a higher prevalence of kyphosis?

```python
# General plot showing the frequency of the Start vertebra number and Kyphosis recurrence status
kyphosis_by_vertebra = df.groupby('Start Vertebrae')['Kyphosis Status'].value_counts(normalize=True).unstack().fillna(0)

# Visualize Kyphosis Range
plt.figure(figsize=(12, 8))
kyphosis_by_vertebra.plot(kind='bar', stacked=True)
plt.title('Range of Kyphosis by the Start Vertebrea Numbers')
plt.xlabel('Start Vertebrea Number')
plt.ylabel('Oran')
plt.legend(title='Kyphosis')
plt.show()
```

### Scale the raw Age column (in months) using both standardization and Normalization. 
- Perform a sanity check.# Normalization is conducted on the 'Age' column to make feature values range from 0 to 1.

```python
# Scale the raw Age column (in months) using both standardization and Normalization. Perform a sanity check.
# Normalization is conducted on the 'Age' column to make feature values range from 0 to 1.

from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Select the age column
age_data = df[['Age']]

# Perform standardization
scaler_standard = StandardScaler()
age_standardized = scaler_standard.fit_transform(age_data)

# Perform normalization
scaler_minmax = MinMaxScaler()
age_normalized = scaler_minmax.fit_transform(age_data)

# Create a new DataFrame to show the results
result_df = df.copy()
result_df['Age_standardized'] = age_standardized
result_df['Age_normalized'] = age_normalized

# Display the first few rows to check the results
print(result_df[['Age', 'Age_standardized', 'Age_normalized']].head())

# Sanity check - The mean of the standardized age should be 0 and the standard deviation should be 1
mean_standardized = age_standardized.mean()
std_standardized = age_standardized.std()

# Sanity check - The min of the normalized age should be 0 and the max should be 1
min_normalized = age_normalized.min()
max_normalized = age_normalized.max()

print(f"\nStandardized Age - Mean: {mean_standardized:.2f}, Std Dev: {std_standardized:.2f}")
print(f"Normalized Age - Min: {min_normalized:.2f}, Max: {max_normalized:.2f}")
```

### Pearson/Spearman Correlations

```python
# Select relevant numerical variables for correlation analysis
numerical_vars = ['Age', 'Number of Vertebrae', 'Start Vertebrae', 'Vertebrae Group Number', 'Kyphosis Status']


# Pearson Correlation
pearson_corr = df[numerical_vars].corr(method='pearson')
print("Pearson Correlation Matrix:")
print(pearson_corr)

# Spearman Correlation
spearman_corr = df[numerical_vars].corr(method='spearman')
print("\nSpearman Correlation Matrix:")
print(spearman_corr)
```

### Pearson Correlation Matrix

```python
# Plot the Pearson Correlation Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Matrix')
plt.show()
```

<div style="border-radius:10px;border:#254E58 solid;padding: 15px;background-color:white;font-size:110%;text-align:left">

<font color=dodgerblue> 💡 Output:The Pearson correlation matrix reveals the following key points: </font>  

1. **Kyphosis Status:**
   - Positively correlated with `Number of Vertebrae` (0.36), indicating that a higher number of affected vertebrae is associated with kyphosis.
   - Negatively correlated with `Start Vertebra` (-0.45) and `Vertebra Group Number` (-0.40), suggesting that kyphosis is more likely to occur when the affected vertebrae are lower in the spine.

2. **Start Vertebra:**
   - Strong positive correlation with `Vertebra Group Number` (0.86), indicating that higher starting vertebra numbers are associated with higher vertebra group numbers.

Overall, the presence of kyphosis is associated with a higher number of affected vertebrae and tends to occur in lower vertebral positions.
