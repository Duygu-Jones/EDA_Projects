- **Dataset:** 'startup_funding.csv'
- **Content:** This dataset contains information about funding provided to startups in India.
- **Rows:** 3044    
- **Columns:** 10   

   INPUT:

**- ``Sr No``** - Serial number for each entry<br>
**- ``Date dd/mm/yyyy``** - The date when the funding was provided, in the format dd/mm/yyyy<br>
**- ``Startup Name``** - The name of the startup that received funding<br>
**- ``Industry Vertical``** - The industry or sector in which the startup operates, such as technology, healthcare, e-commerce, etc.<br>
**- ``SubVertical``** - A more specific category within the industry, providing additional details about the startup's focus<br>
**- ``City  Location``** - The city or location where the startup is based<br>
**- ``Investors Name``** - The names of the investors or venture capital firms that provided the funding<br>
**- ``InvestmentnType``** - The type of investment, such as Seed Funding, Series A, Series B, etc.<br>
**- ``Amount in USD``** - The amount of funding received by the startup, typically expressed in US dollars<br>
**- ``Remarks``** - Any additional remarks or notes about the funding round or the startup<br>

#### `Amount_USD`

```python
# Replace commas(,) in the 'amount_USD' column
df['amount_USD'] = df['amount_USD'].apply(lambda x: str(x).replace(',', ''))


# Correct invalid values in the 'amount_USD' column
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("undisclosed", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("Undisclosed", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("unknown", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("14342000+", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("\\\\xc2\\\\xa010000000", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("\\\\xc2\\\\xa05000000", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("\\\\xc2\\\\xa019350000", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("\\\\xc2\\\\xa0600000", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("\\\\xc2\\\\xa020000000", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("\\\\xc2\\\\xa0N/A", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("\\\\xc2\\\\xa016200000", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("\\\\xc2\\\\xa0685000", "0"))
df['amount_USD'] = df['amount_USD'].apply(lambda x : str(x).replace("nan", "0"))
```

```python
# Convert to numeric
df['amount_USD'] = pd.to_numeric(df['amount_USD'], errors='coerce')

# Replace 0 values in 'amount_USD' with NaN
df['amount_USD'] = df['amount_USD'].replace(0, np.nan)

# Replace NaN values with the mean
df['amount_USD'].fillna(df['amount_USD'].mean(), inplace=True)
```

###  How Does the Funding Ecosystem changes with respect to Time? 

```python
# Aggregate data by year and month using existing columns
monthly_funding = df.groupby(['year', 'month']).agg(
    total_funding=('amount_USD', 'sum'),
    number_of_investments=('amount_USD', 'count'),
    average_funding=('amount_USD', 'mean')
).reset_index()

# Create a 'year_month' column for plotting
monthly_funding['year_month'] = pd.to_datetime(monthly_funding[['year', 'month']].assign(day=1))
```

```python
# Plot total funding amount over time
plt.figure(figsize=(10,5))
plt.plot(monthly_funding['year_month'], monthly_funding['total_funding'], marker='o', linestyle='-')
plt.title('Total Funding Amount Over Time')
plt.xlabel('Time')
plt.ylabel('Total Funding Amount (USD)')
plt.grid(True)
plt.show()
```

<font color=dodgerblue>  Output:  Histogram of Bike Share Distribution</font>  

- **Fluctuations**: Noticeable fluctuations in total funding amounts between 2015 and 2020.
- **Peak**: Peak funding amount was observed in 2019.
- **Low Activity**: Low funding activity in late 2018 and early 2019.

# Does Location also play a role, In determining the Growth of a Startup? (Bir Startup'n Bymesini Belirlemede Lokasyonun da Rol Var m?)

```python
# Veri ykleme
file_path = ('startup_funding.csv')
data = pd.read_csv(file_path)

# Kolon isimlerini dzenleme
data.rename(columns={'Date dd/mm/yyyy': 'Date', 'City  Location': 'City'}, inplace=True)

# Tarih kolonunu datetime formatna evirme
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y', errors='coerce')

# Tarih dnm baarsz olan satrlar kaldrma
data = data.dropna(subset=['Date'])

# 'Amount in USD' kolonundaki virglleri kaldrma ve float tipine evirme
data['Amount in USD'] = data['Amount in USD'].str.replace(',', '')
data = data[pd.to_numeric(data['Amount in USD'], errors='coerce').notnull()]
data['Amount in USD'] = data['Amount in USD'].astype(float)

# Temizlenmi veriyi kontrol etme
data.head()
```

```python
# ehir baznda toplam yatrm miktarn hesaplama
city_funding = data.groupby('City')['Amount in USD'].sum().sort_values(ascending=False)
city_funding.head()
```

```python
# Grafik oluturma
plt.figure(figsize=(12, 6))
city_funding.plot(kind='bar', color='blue')
plt.title('Funding Amount by City')
plt.xlabel('City')
plt.ylabel('Total Funding Amount (USD)')
plt.xticks(rotation=90)
plt.show()
```

```python


# Gerekli stunlar se
city_funding = dataset.groupby('City')['Amount in USD'].count().reset_index()
city_funding.columns = ['City', 'Number of Fundings']

# ehir baznda toplam fon miktarn hesapla
dataset['Amount in USD'] = dataset['Amount in USD'].astype(str).str.replace(',', '').astype(float)
city_total_funding = dataset.groupby('City')['Amount in USD'].sum().reset_index()
city_total_funding.columns = ['City', 'Total Funding (USD)']

# ehir baznda fonlanan startup says ve toplam fon miktarn birletir
city_stats = pd.merge(city_funding, city_total_funding, on='City')

# ehir baznda fonlanan startup says ve toplam fon miktarn grselletir
fig, ax1 = plt.subplots(figsize=(14, 7))
ax2 = ax1.twinx()

sns.barplot(x='City', y='Number of Fundings', data=city_stats, ax=ax1, palette='viridis')
sns.lineplot(x='City', y='Total Funding (USD)', data=city_stats, ax=ax2, color='red', marker='o')

ax1.set_xlabel('City')
ax1.set_ylabel('Number of Fundings', color='blue')
ax2.set_ylabel('Total Funding (USD)', color='red')
ax1.set_title('ehir Baznda Fonlanan Startup Says ve Toplam Fon Miktar')
ax1.tick_params(axis='x', rotation=90)
ax1.grid()

plt.show()

```

# Who plays the main role in Indian Startups Ecosystem? (Hint Startup Ekosistemi'nde ana rol kim oynuyor?)


Hint Startup Ekosistemi'nde ana rol oynayanlar belirlemek iin birka farkl analiz yaplabilir. Genellikle bu tr bir soruya cevap vermek iin aadaki faktrlere baklr:

Yatrmclar: Hangi yatrmclar en ok yatrm yapyor?
Sektrler: Hangi sektrler en ok yatrm alyor?
ehirler: Hangi ehirler en ok yatrm ekiyor?
Bu faktrleri analiz ederek, Hint startup ekosisteminde ana rol kimlerin oynadn belirleyebiliriz. Aadaki admlarla bu analizleri gerekletirebiliriz:

Yatrmc Analizi: En ok yatrm yapan yatrmclar belirleyin.
Sektr Analizi: En ok yatrm alan sektrleri belirleyin.
ehir Analizi: En ok yatrm alan ehirleri belirleyin.

Sonu
Yukardaki analizler ile Hint startup ekosisteminde ana rol oynayan yatrmclar, sektrleri ve ehirleri belirleyebilirsiniz. Bu analizler, hangi yatrmclarn en ok yatrm yaptn, hangi sektrlerin en ok yatrm aldn ve hangi ehirlerin en ok yatrm ektiini gsterir.

```python

# Yatrmc analizi
investor_funding = data['Investors Name'].value_counts().reset_index()
investor_funding.columns = ['Investor', 'Number of Investments']

# En ok yatrm yapan ilk 10 yatrmc
top_investors = investor_funding.head(10)

plt.figure(figsize=(14, 7))
plt.barh(top_investors['Investor'], top_investors['Number of Investments'], color='skyblue')
plt.xlabel('Number of Investments')
plt.ylabel('Investor')
plt.title('Top 10 Investors in Indian Startup Ecosystem')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Sektr analizi
sector_funding = data['Industry Vertical'].value_counts().reset_index()
sector_funding.columns = ['Industry Vertical', 'Number of Investments']

# En ok yatrm alan ilk 10 sektr
top_sectors = sector_funding.head(10)

plt.figure(figsize=(14, 7))
plt.barh(top_sectors['Industry Vertical'], top_sectors['Number of Investments'], color='lightgreen')
plt.xlabel('Number of Investments')
plt.ylabel('Industry Vertical')
plt.title('Top 10 Sectors in Indian Startup Ecosystem')
plt.gca().invert_yaxis()
plt.show()
```

```python
# ehir analizi
city_funding = data['City'].value_counts().reset_index()
city_funding.columns = ['City', 'Number of Investments']

# En ok yatrm alan ilk 10 ehir
top_cities = city_funding.head(10)

plt.figure(figsize=(14, 7))
plt.barh(top_cities['City'], top_cities['Number of Investments'], color='salmon')
plt.xlabel('Number of Investments')
plt.ylabel('City')
plt.title('Top 10 Cities in Indian Startup Ecosystem')
plt.gca().invert_yaxis()
plt.show()
```

# What are the different Types of Funding for Startups?(Startup'lar iin farkl Finansman Trleri nelerdir?)

```python

# Finansman trlerini gruplandr ve saysn hesapla
financing_types = dataset['InvestmentnType'].value_counts().reset_index()
financing_types.columns = ['Investment Type', 'Count']

# Finansman trlerini grselletir
plt.figure(figsize=(14, 7))
sns.barplot(x='Count', y='Investment Type', data=financing_types, palette='viridis')
plt.title('Startup\'lar iin Farkl Finansman Trleri')
plt.xlabel('Count')
plt.ylabel('Investment Type')
plt.grid()
plt.show()
```

```python


# Finansman trlerini gruplandr ve saysn hesapla
financing_types = dataset['InvestmentnType'].value_counts().reset_index()
financing_types.columns = ['Investment Type', 'Count']

# En ok rol oynayan ilk 6 finansman tr
top_financing_types = financing_types.head(6)

# Dier finansman trlerini gruplandr
other_financing_types = financing_types[6:]
other_examples = ', '.join(other_financing_types['Investment Type'].head(3))
other_summary = pd.DataFrame([{
    'Investment Type': f'Other (e.g., {other_examples})',
    'Count': other_financing_types['Count'].sum()
}])

# lk 6 tr ve dierleri birletir
final_financing_types = pd.concat([top_financing_types, other_summary])

# Finansman trlerini grselletir
plt.figure(figsize=(14, 7))
sns.barplot(x='Count', y='Investment Type', data=final_financing_types, palette='viridis')
plt.title('Startup\'lar iin Farkl Finansman Trleri')
plt.xlabel('Count')
plt.ylabel('Investment Type')
plt.grid()
plt.show()
```
