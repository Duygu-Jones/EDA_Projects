## <a id='toc2_3_'></a>[Sutunlarin Duzenlenmesi:](#toc0_)
Genel olark datasetinin yapisinin inceledik bundan sonraki islemleri kolaylastirmak icin Sutun ve isimlerini standartlastiralim 
- Null cogunlukta ve anlamsiz sutunlarin tespit edilmesi ve cikarilmasi

```python
# ========User-Defined-Fonction:========================

# ========Summary of Object Columns=================
def object_summary(df):
    obs = df.shape[0]
   
    object_df = df.select_dtypes(include='object')
    summary_df = pd.DataFrame({
        'Dtype': object_df.dtypes,
        'Counts': object_df.apply(lambda x: x.count()),
        'Nulls': object_df.apply(lambda x: x.isnull().sum()),
        'NullPercent': (object_df.isnull().sum() / obs) * 100,
        'Top': object_df.apply(lambda x: x.mode()[0] if not x.mode().empty else '-'),
        'Frequency': object_df.apply(lambda x: x.value_counts().max() if not x.value_counts().empty else '-'),
        'Uniques': object_df.apply(lambda x: x.unique().shape[0]),
        'UniqueValues': object_df.apply(lambda x: list(x.unique()) if x.nunique() <= 50 else '-')
    })
    
    # Format 'NullPercent' to show percentages with two decimal points and a '%' sign
    summary_df['NullPercent'] = summary_df['NullPercent'].map("{:.2f}%".format)
    
    print('Categorical Features Summary:')
    print('___________________________\nData Types:')
    print(summary_df['Dtype'].value_counts())
    print('___________________________')
    
    
    return summary_df


# =============Summary of Numeric Features================= 
def numeric_summary(df):
    obs = df.shape[0]
    
    numeric_df = df.select_dtypes(include='number')
    summary_df = pd.DataFrame({
        'Dtype': numeric_df.dtypes, 
        'Counts': numeric_df.apply(lambda x: x.count()), 
        'Nulls': numeric_df.apply(lambda x: x.isnull().sum()), 
        'NullPercent': (numeric_df.isnull().sum() / obs) * 100,
        'Min': numeric_df.min(),
        'Max': numeric_df.max(),
        'Uniques': numeric_df.apply(lambda x: x.unique().shape[0]),
        'UniqueValues': numeric_df.apply(lambda x: list(x.unique()) if x.nunique() <= 10 else '-')
    })
    
    print('___________________________\nData Types:')
    print(summary_df['Dtype'].value_counts())
    print('___________________________')
    print('Numreical Features Summary:')
    
    return summary_df
```

### <a id='toc6_2_1_'></a>[Hangi ürünler en çok satılıyor? Pizza ve içecek kategorilerinde popülarite sıralaması nasıl?](#toc0_)

```python
urunler_list = df['urunler'].unique()
print('Urunler sutunundaki farkli deger sayisi:', len(urunler_list))
urunler_list
```

```python
# 1. Ürünleri ayıklama fonksiyonu
def urunleri_ayikla(df):
    urunler_listesi = []
    
    for index, row in df.iterrows():
        siparis_no = row['siparis_no']
        urunler_str = str(row['urunler'])
        
        # Virgülle ayır
        urun_listesi = urunler_str.split(', ')
        
        for urun in urun_listesi:
            # Miktarı ve ürün adını ayır
            parcalar = urun.split(' ', 1)
            
            if len(parcalar) >= 2:
                try:
                    miktar = int(parcalar[0])
                    urun_adi = parcalar[1]
                    
                    # Ürün adını temizle ve kategorize et
                    urun_temiz = urun_kategorize(urun_adi)
                    
                    if urun_temiz:  # Eğer kategorize edilebildiyse
                        urunler_listesi.append({
                            'siparis_no': siparis_no,
                            'urun': urun_temiz,
                            'miktar': miktar
                        })
                except ValueError:
                    continue
    
    return pd.DataFrame(urunler_listesi)

# 2. Ürün kategorilendirme fonksiyonu
def urun_kategorize(urun_adi):
    urun_adi = urun_adi.lower()
    
    # Pizza türleri
    if 'roma' in urun_adi:
        return 'Roma Pizza'
    elif 'milano' in urun_adi:
        return 'Milano Pizza'
    elif 'torino' in urun_adi:
        return 'Torino Pizza'
    elif 'verona' in urun_adi:
        return 'Verona Pizza'
    elif 'margherita' in urun_adi:
        return 'Margherita Pizza'
    elif 'napolitan' in urun_adi or 'neapolitan' in urun_adi:
        return 'Napolitan Pizza'
    
    # İçecekler
    elif 'coca-cola' in urun_adi or 'coca cola' in urun_adi:
        return 'Coca-Cola'
    elif 'sprite' in urun_adi:
        return 'Sprite'
    elif 'fanta' in urun_adi:
        return 'Fanta'
    elif 'lipton' in urun_adi:
        return 'Lipton'
    elif 'su' in urun_adi and 'cl' in urun_adi:
        return 'Su'
    elif 'poşet' in urun_adi:
        return 'Poşet'
    
    return None  # Kategorize edilemeyen ürünler

# 3. Ürünleri ayıkla
urunler_df = urunleri_ayikla(df)
urunler_df.head(10)
```

```python
# 4. Ürün satış miktarlarını hesapla
urun_satis = urunler_df.groupby('urun')['miktar'].sum().sort_values(ascending=False)

# 5. Grafik oluştur
plt.figure(figsize=(12, 6))
bars = plt.bar(range(len(urun_satis)), urun_satis.values, color='skyblue', alpha=0.8)

# Her urunun miktar goster 
for i, (bar, value) in enumerate(zip(bars, urun_satis.values)):
   plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1, 
            str(value), ha='center', va='bottom', fontsize=10, fontweight='bold')
   
plt.title('Ürünlere Göre Satış Miktarları', fontsize=16, fontweight='bold')
plt.xlabel('Ürünler', fontsize=12)
plt.ylabel('Satış Miktarı', fontsize=12)

# X ekseni etiketleri
plt.xticks(range(len(urun_satis)), urun_satis.index, rotation=45, ha='right')

# Y ekseni için grid
plt.grid(axis='y', alpha=0.3)

# Layout ayarla
plt.tight_layout()

# Grafik göster
plt.show()
```

# <a id='toc7_'></a>[URUN BAZLI KAR ANALIZI](#toc0_)

```python
# Urulerin ayristirildigi datasete gerekli diger sutunlari da ekleyelim:

df_filtered = df[['siparis_no', 'toplam_tutar', 'tahmini_kazanc']]

# Merge
analiz_df = urunler_df.merge(df_filtered, on='siparis_no')
analiz_df.head(3)
```

```python
# Kar paylasim hesaplamasi:

# Her siparişin toplam karını, ürün miktarlarına orantılı olarak paylaştırır
siparis_toplam = analiz_df.groupby('siparis_no')['miktar'].transform('sum')
analiz_df['urun_kar'] = analiz_df['tahmini_kazanc'] * (analiz_df['miktar'] / siparis_toplam)

# Ürün bazlı karlılık
karlillik = analiz_df.groupby('urun').agg({
    'miktar': 'sum',
    'urun_kar': 'sum'
}).round(2)
karlillik['birim_kar'] = (karlillik['urun_kar'] / karlillik['miktar']).round(2)
karlillik = karlillik.sort_values('urun_kar', ascending=False)

print("📊 En Karlı Ürünler:")
karlillik.head(10)
```

```python
kupon_sayilar = kupon_df['kupon_kategori'].value_counts()
kupon_yuzde = kupon_df['kupon_kategori'].value_counts(normalize=True) * 100

print("Kupon Kullanım İstatistikleri:")
for kategori, sayi in kupon_sayilar.items():
    print(f"{kategori}: {sayi} sipariş ({kupon_yuzde[kategori]:.1f}%)")

# Grafik 1: Kupon Kullanım Oranları
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
colors = ['#FF6B6B', '#4ECDC4']

# Pasta grafiği
axes[0].pie(kupon_sayilar.values, labels=kupon_sayilar.index, 
           autopct='%1.1f%%', colors=colors, startangle=90)
axes[0].set_title('Kupon Kullanım Oranları', fontweight='bold')

# Bar grafiği
bars = axes[1].bar(kupon_sayilar.index, kupon_sayilar.values, color=colors, alpha=0.7)
for bar, value, pct in zip(bars, kupon_sayilar.values, kupon_yuzde.values):
    axes[1].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
                f'{value}\n({pct:.1f}%)', ha='center', va='bottom', fontweight='bold')
axes[1].set_title('Kupon Kullanım Dağılımı', fontweight='bold')
axes[1].set_ylabel('Sipariş Sayısı')

plt.tight_layout()
plt.show()
```

- Kupon benimseme oranı oldukça yüksek - müşteriler fiyat duyarlı
- Kupon stratejisi başarılı ve müşteri davranışını etkiliyor
- Kuponsuz müşteri segmenti küçük (%25.8) - bu gruba özel stratejiler geliştirilebilir

-> Öneri: Kupon kullanımı dominant olduğu için kupon kampanyalarına odaklanmak mantıklı, ancak kuponsuz müşterilerin neden kupon tercih etmediğini araştırmak da fırsatlar yaratabilir.

### Kupon kampanyaları karlılığı artırıyor mu, ROI pozitif mi?

```python
# Kupon performans analizi
kupon_analiz = kupon_df.groupby('kupon_kategori').agg({
    'toplam_tutar': 'mean',
    'tahmini_kazanc': ['mean', 'sum'],
    'toplam_kupon': ['mean', 'sum']
}).round(2)

print(f"\nKupon Performans Analizi:")
display(kupon_analiz)

# ROI hesaplama
kupon_maliyet = kupon_df[kupon_df['kupon_var']]['toplam_kupon'].sum()
kupon_kar = kupon_df[kupon_df['kupon_var']]['tahmini_kazanc'].sum()
roi = ((kupon_kar - kupon_maliyet) / kupon_maliyet * 100) if kupon_maliyet > 0 else 0

print(f"\nROI Analizi:")
print(f"Kupon Maliyeti: {kupon_maliyet:.2f}₺")
print(f"Kuponlu Kar: {kupon_kar:.2f}₺")
print(f"ROI: {roi:.1f}%")
```

ROI (Return on Investment):
- Yatırım geri dönüş oranı - kupon kampanyalarına harcanan paranın ne kadari kar getiriyor?
    - (Elde Edilen Kar - Yatırım Maliyeti) / Yatırım Maliyeti × 100
- > Her 100₺ kupon harcamasından 14.5₺ net kar elde ediliyor

Not: 
- ROI pozitif fakat Kuponsuz müşteriler daha yüksek kar marjı sağlıyor 
- Kupon kullanımı kar marjını düşürüyor ama satış hacmini artırıyor
- > Kupon stratejisini sürdürmek mantıklı (%14.5 ROI kabul edilebilir) ancak kuponsuz premium müşteri segmentini de büyütmeye odaklanmak karlılığı artırabilir.

```python
# Kupon vs İndirim Karlilik Karşılaştırması

def heatmap_table():
    import seaborn as sns
    
    # Pivot table oluştur
    pivot_data = []
    for i, row in karsilastirma_df.iterrows():
        strategy_type = 'Kupon' if 'Kupon' in row['Strateji'] else 'İndirim'
        usage = 'Kullanan' if 'Kullanan' in row['Strateji'] else 'Kullanmayan'
        pivot_data.append([strategy_type, usage, row['Ortalama_Kar']])
    
    pivot_df = pd.DataFrame(pivot_data, columns=['Strateji', 'Kullanım', 'Kar'])
    pivot_table = pivot_df.pivot(index='Kullanım', columns='Strateji', values='Kar')
    
    plt.figure(figsize=(8,6))
    sns.heatmap(pivot_table, annot=True, cmap='RdYlGn', fmt='.1f', 
                cbar_kws={'label': 'Ortalama Kar (₺)'})
    plt.title('Strateji Kar Matrisi', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

heatmap_table()
```
